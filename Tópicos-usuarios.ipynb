```python
from nltk.tokenize import TweetTokenizer

tknzr = TweetTokenizer()
```

```python
stopwords
```

```python
with open(stopwords, encoding = 'utf8') as sw_file:
    stopwords = [word[:-1].lower() for word in sw_file.readlines()]
    extras = ['\n', '\t', '\r', 'rt']
    stopwords.extend(extras)
```

```python
def filter(list):
    return [i for i in list if i not in stopwords]
```

```python
def joinList(list):
    return ' '.join(list)
```

```python
regex_dict = {
    'punct': re.compile(r'[^@#\w_]'),
    'menciones': re.compile(r'@[\w_]+\b'),
    'hashtags': re.compile(r'#[\w_]+\b'),
    'urls': re.compile(r'(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?')
}
```

```python
tweets = data['Texto'].str.replace('RT ', '')
tweets = tweets.str.lower()
tweets = tweets.str.replace(regex_dict['urls'], '')
tweets = tweets.str.replace(regex_dict['punct'], ' ')
tweets = tweets.str.replace(regex_dict['hashtags'], '')
tweets= tweets.str.replace(regex_dict['menciones'], '')
tweets = tweets.str.replace(r'\W{2,}', ' ')
tweets = tweets.str.replace(r'[0-9]', '')
tweets = tweets.apply(tknzr.tokenize)
tweets = tweets.apply(filter)
tweets = tweets.apply(joinList)
print(tweets)
```

```python
from bertopic import BERTopic
```

```python
topic_model = BERTopic(language="multilingual", min_topic_size=15, nr_topics=20)#, calculate_probabilities = True) #, verbose=True)
#topic_model = BERTopic.load("topicos_25")
topics, probs = topic_model.fit_transform(tweets)
```

```python
freq = topic_model.get_topic_info(); freq.head(60)
```

```python
for i in range(1, 21):    
    topic_nr = freq.iloc[i]["Topic"]
    print(f'Tópico {i-1}:')
    print(topic_model.get_topic(topic_nr)) 
```
