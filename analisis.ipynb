```python
import re
import pandas as pd
from datetime import datetime
from nltk.tokenize import TweetTokenizer
tknzr = TweetTokenizer()
from config_1 import stopwords
```

```python
from nltk.tokenize import TweetTokenizer

tknzr = TweetTokenizer()
```

```python
with open(stopwords, encoding = 'utf8') as sw_file:
    stopwords = [word[:-1].lower() for word in sw_file.readlines()]
    extras = ['\n', '\t', '\r', 'rt']
    stopwords.extend(extras)
```

```python
def filter(list):
    return [i for i in list if i not in stopwords]
```

```python
def joinList(list):
    return ' '.join(list)
```

```python
regex_dict = {
    'punct': re.compile(r'[^@#\w_]'),
    'menciones': re.compile(r'@[\w_]+\b'),
    'hashtags': re.compile(r'#[\w_]+\b'),
    'urls': re.compile(r'(http|ftp|https)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?')
}
```

```python
tweets = data_1['Tweet'].str.replace('RT ', '')
tweets = tweets.str.lower()
tweets = tweets.str.replace(regex_dict['urls'], '')
tweets = tweets.str.replace(regex_dict['punct'], ' ')
tweets = tweets.str.replace(regex_dict['hashtags'], '')
tweets= tweets.str.replace(regex_dict['menciones'], '')
tweets = tweets.str.replace(r'\W{2,}', ' ')
tweets = tweets.str.replace(r'[0-9]', '')
tweets = tweets.apply(tknzr.tokenize)
tweets = tweets.apply(filter)
tweets = tweets.apply(joinList)
tweets = tweets.sample(frac=1).reset_index(drop=True)
tweets = tweets[0:1500]
```

## Hastags

```python
hastags = data_1['Hastags']
```

```python
hastags
```

```python
import ast
x = ast.literal_eval(hastags[0])
```

```python
x[0]
```

```python
hastags_1 = []
for i in range(len(hastags)):
    hastags[i] = ast.literal_eval(hastags[i])
    if hastags[i] == []:
        pass
    else:
        hastags_1 = hastags_1 + hastags[i]
        
```

```python
import collections
```

```python
counter=collections.Counter(hastags_1)
print(counter)
```

```python
relevantes = dict((k, v) for k, v in counter.items() if v >= 3)
```

```python
relevantes= {k: v for k, v in sorted(relevantes.items(), key=lambda item: item[1], reverse=True)}
```

```python
import numpy as np
import matplotlib.pyplot  as plt
```

```python
vals = np.fromiter(relevantes.values(), dtype=float)
y_pos = np.arange(len(relevantes.keys()))
```

```python
cmap = plt.cm.Paired
colors = cmap(np.arange(len(relevantes.keys())) % cmap.N)
```

```python
plt.rcdefaults()

fig, ax = plt.subplots(figsize=(10,10))

# Example data
#people = ('Tom', 'Dick', 'Harry', 'Slim', 'Jim')
#y_pos = np.arange(len(people))
#performance = 3 + 10 * np.random.rand(len(people))
#error = np.random.rand(len(people))

ax.barh(y_pos, vals, align='center', color =colors)
ax.set_yticks(np.arange(len(relevantes)))
ax.set_yticklabels(relevantes.keys())
ax.invert_yaxis()  # labels read top-to-bottom
#ax.set_color(colors)
ax.set_xlabel('Recuento')
ax.set_title('Hastaghs')
plt.savefig('Hastags.png')

```
